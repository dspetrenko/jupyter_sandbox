```python
import pandas as pd
```

```python
weeks  = [ 'W2018{0:02}'.format(week) for week in range(1, 53)]
df = pd.DataFrame({'YW':weeks})
```

```python
df['Year'] = df['YW'].apply(lambda x: int(x[1:5]))
df['Week'] = df['YW'].apply(lambda x: int(x[5:]))
```

```python
df.head()
```

```python
df.tail()
```

```python
df['SO Date'] = pd.to_datetime(df.Year.astype(str), format='%Y') + \
             pd.to_timedelta(df.Week.mul(7).astype(str) + ' days')
df['SO Week'] = df['SO Date'].dt.week
```

```python
df.head()
```

```python
df.tail()
```

You can see that 'SO Date' contain monday of next week.
More correct way placed below (columns name contains prefix 'CW'):

```python
df['CW Date'] = pd.to_datetime(df['Year'].astype(str), format='%Y') + \
                pd.to_timedelta((df['Week'].mul(7) - 7).astype(str) + ' days')
df['CW Week'] = df['CW Date'].dt.week
```
